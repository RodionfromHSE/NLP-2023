```python
from semantic_searcher import SemanticSearcher

ss = SemanticSearcher()
```

```python
query = "Main city of Japan"
messages = [
    "Tokyo is the capital of Japan.",
    "Tokyo is the largest city of Japan.",
    "Berlin is the capital of Germany.",
    "I love pizza.",
]

similar_messages, scores = ss(query, messages, return_scores=True)
for msg, score in zip(similar_messages, scores):
    print(f"{msg} (score: {score})")
```
