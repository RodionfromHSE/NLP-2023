```python
import json

with open('arxivData.json', 'r') as f:
    data = json.load(f)
```

```python
# make txt file Title: ...\n Summary: ...\n

texts = []
for i in range(len(data)):
    texts.append('Title: ' + data[i]['title'] + '\nSummary: ' + data[i]['summary'])
```

```python
with open('arxivData.txt', 'w') as f:
    for text in texts:
        f.write(text + '\n')
```
